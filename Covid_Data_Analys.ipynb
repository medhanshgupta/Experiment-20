```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/covid_19_data.csv")
data.head()
```

```python
data = data.drop(['SNo', 'Last Update'], axis=1)
data.head()
```

```python
data.info()
```

```python
data['ObservationDate'] = data['ObservationDate'].astype('datetime64[ns]')
data['Confirmed'] = data['Confirmed'].fillna(0).astype('int64')
data['Deaths'] = data['Deaths'].fillna(0).astype('int64')
data['Recovered'] = data['Recovered'].fillna(0).astype('int64')
data.info()
```

```python
### Active Case

data['Active'] = data['Confirmed'] - data['Deaths'] - data['Recovered']
data.head(50)

data.iloc[50:100]
```

```python
### last/current date
data['ObservationDate'].max()
```

```python
### Fetch the latest date data

latest_data = data[data['ObservationDate'] == data['ObservationDate'].max()]
latest_data.head()
```

```python
latest_data.shape
```

```python
latest_data['Country/Region'].value_counts()
```

```python
latest_data['Country/Region'].nunique()
```

```python
latest_data['Country/Region'].unique()
```

```python
## Count of cases for each country

countries = latest_data.groupby('Country/Region')[["Confirmed" , "Deaths" , "Recovered" , "Active"]].sum()

countries = countries.reset_index()
```

```python
countries
```

```python
# Cases in India

countries[countries['Country/Region'] == "India"]
```

```python
### Plot on World Map
```

```python
import plotly.express as px
```

```python
world_map = px.choropleth(countries,locations="Country/Region",locationmode="country names",color="Confirmed",color_continuous_scale="reds",range_color=[0.1, countries['Confirmed'].max()])
world_map
```

```python
### Top countries having maximum confirmed and recovered cases
```

```python
top = latest_data.groupby("Country/Region")[['Confirmed' , 'Recovered']].sum().reset_index()
top = top.sort_values(by = 'Confirmed' , ascending = False).reset_index(drop = True)
top.head()
```

```python
top_20 = top.head(20)
top_20
```

```python
india = data[data['Country/Region']=="India"]
india
```

```python
# Find total number of states

india['Province/State'].nunique()
```

```python
india['Province/State'] = india['Province/State'].fillna('mode')
india.head()
```

```python
india_latest_data = india[india['ObservationDate'] ==india['ObservationDate'].max()]
india_latest_data
```

```python
top = india_latest_data.groupby("Province/State")[['Confirmed' , 'Recovered']].sum().reset_index()
top = top.sort_values(by = 'Confirmed' , ascending = False).reset_index(drop = True)
top.head(20)
```

```python
top['Confirmed'].max()
```

```python
top[top['Confirmed'] == top['Confirmed'].max()]
```

```python
import json
import plotly.express as px

# Path to your GeoJSON file for Indian states
geojso_path = '/content/india.json'

india_states_geojson = None
try:
    with open(geojso_path, 'r') as f:
        india_states_geojson = json.load(f)
    print(f"Successfully loaded GeoJSON from {geojso_path}")
except FileNotFoundError:
    print(f"Error: GeoJSON file not found at {geojso_path}. Please upload it or specify the correct path.")
    print("You can often find India state GeoJSON files on GitHub, e.g., 'https://raw.githubusercontent.com/geohacker/india/master/state/india_state.geojson'")

if india_states_geojson:
    # It's crucial that the 'Province/State' names in your `top` DataFrame
    # exactly match the names in the GeoJSON file's 'properties' for `featureidkey`.
    # Common keys in GeoJSON for state names are 'NAME_1', 'ST_NM', 'NAME', 'state_name', etc.
    # Adjust 'properties.st_nm' if your GeoJSON uses a different key.

    # Example: Check GeoJSON properties to determine correct featureidkey
    if 'features' in india_states_geojson and len(india_states_geojson['features']) > 0:
        print("\nKeys in GeoJSON properties for the first feature:")
        print(india_states_geojson['features'][0]['properties'].keys())

    fig = px.choropleth(
        top,
        geojson=india_states_geojson,
        locations='Province/State',  # Column in your DataFrame that contains state names
        featureidkey='properties.st_nm', # Path to the state name in the GeoJSON properties (changed from NAME_1 to st_nm)
        color='Confirmed',           # Column to color the map by
        hover_name='Province/State', # Display state name on hover
        color_continuous_scale="Plasma",
        projection="mercator",
        title='COVID-19 Confirmed Cases in Indian States'
    )

    fig.update_geos(fitbounds="locations", visible=False) # Automatically zoom to India
    fig.show()
else:
    print("Cannot plot map without valid GeoJSON data.")
```
